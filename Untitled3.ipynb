```python
import pandas as pd
import numpy as np
#sample data set
data={'age':[25,27,np.nan,100,30,27],
'salary':[5000,52000,51000,50500,52000,52000],
'city':['A','B','A','A','A', np.nan]
}
df=pd.DataFrame(data)
print('original data:')
print(df)
#1.handeling missing values
df['age']=df['age'].fillna(df['age'].mean())
print("\n after handeling missing value")
print(df)
#2.noise deduction(removing outlier)
Q1=df['age'].quantile(0.25)
Q3=df['age'].quantile(0.75)
IQR=Q3-Q1
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
df=df[(df['age']>=lower_limit)&
      (df['age']<=upper_limit)]
print("\n after noise deduction")
print(df)

df=df.drop_duplicates('age')
print("\n after removing duplicates")
print(df)
#3.Identifying & removing data redudency(duplicater)
df=df.drop_duplicates()
print("\n after removing data redudant(duplicate)records:")
print(df)
```
